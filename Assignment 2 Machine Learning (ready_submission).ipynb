In this exercise, we ask you to predict wine quality in a multi-class classification setup using a LogisticRegression classifier. In less than 50 words, present the potential advantages and drawbacks of viewing this problem as a multi-class classification task.

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
dataredwine = pd.read_csv('/Users/yessinradouane/Desktop/Uni/ML/Assignment1/winequality/winequality-red.csv', sep=';')
datawhitewine = pd.read_csv('/Users/yessinradouane/Desktop/Uni/ML/Assignment1/winequality/winequality-white.csv', sep=';')
```

```python
from collections import Counter
from sklearn.datasets import make_classification
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

features = list(dataredwine.keys())
features.remove('quality')
quality_W = dataredwine['quality']
 
def split_data(data):
    X = data.loc[:, features]
    y = data.loc[:, ['quality']]
 
    X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 6)
 
    return X_train, X_test, y_train, y_test

def fit_regression (x,y):
    model = LogisticRegression(multi_class = 'multinomial').fit(x,y)
    return model

def regression_summary (model,xtrain,ytrain):

    accuracy = model.score(xtrain,ytrain)
    # print(accuracy)
    return accuracy
```

```python
# redwine
X_train, X_test, y_train, y_test = split_data(dataredwine)
# X_train.shape
model = fit_regression(X_train, y_train)
print(regression_summary(model, X_train, y_train))

#whitewine
X_train2, X_test2, y_train2, y_test2 = split_data(datawhitewine)
# X_train.shape
model2 = fit_regression(X_train2, y_train2)
print(regression_summary(model2, X_train2, y_train2))
```

```python
# scale, classifier into pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler 
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import *
from datetime import datetime
from sklearn.metrics import classification_report


def create_pipeline(scalar, classifier):

    start=datetime.now()



    #excluding ID from model
    features = list(train_dummied.keys())
    features.remove('y_yes')
    features.remove('ID')

    #defining data set
    X_trn = train_dummied.loc[:, features]
    y_trn = train_dummied.loc[:, ['y_yes']]
    X_tst = test_dummied.loc[:, features]
    y_tst = test_response_dummied.loc[:, ['y_yes']]


    if classifier == "LinearSVC":

        #defining pipeline
        pipe = make_pipeline(scalar, LinearSVC())

        #get param grid
        c_grid = {"linearsvc__C": np.logspace(-3,3,10)}

        #execute GridSearch
        gridcv = GridSearchCV(pipe, c_grid, cv=5, return_train_score=True)

        #fit data
        gridcv.fit(X_trn, y_trn)

        best_param = gridcv.best_params_

        #y_predicted 
        y_pred = gridcv.predict(X_tst)
        print(classification_report(y_tst, y_pred))


        print(50*"-")
        print(f"{str(scalar)}, {str(classifier)}()")
        print(f"[+] Best parameter: {best_param}\n[+] Test Score: {gridcv.score(X_tst, y_tst)}\n[+] Train Score: {gridcv.score(X_trn, y_trn)}\n[+] Total runtime: {datetime.now()-start}")


    if classifier == "LogisticRegression":

        #defining pipeline
        pipe = make_pipeline(scalar, LogisticRegression())

        #get param grid
        c_grid = {"logisticregression__C": np.logspace(-3,3,10)}

        #execute GridSearch
        gridcv = GridSearchCV(pipe, c_grid, cv=5, return_train_score=True)

        #fit data
        gridcv.fit(X_trn, y_trn)

        best_param = gridcv.best_params_

        #y_predicted 
        y_pred = gridcv.predict(X_tst)
        print(classification_report(y_tst, y_pred))

        print(50*"-")
        print(f"{str(scalar)}, {str(classifier)}()")
        print(f"[+] Best parameter: {best_param}\n[+] Test Score: {gridcv.score(X_tst, y_tst)}\n[+] Train Score: {gridcv.score(X_trn, y_trn)}\n[+] Total runtime: {datetime.now()-start}")


    if classifier == "KNN":
    
        #defining pipeline
        pipe = make_pipeline(scalar, KNeighborsClassifier())

        #get param grid
        n_grid = {"kneighborsclassifier__n_neighbors": range(1,15)}

        #execute GridSearch
        gridcv = GridSearchCV(pipe, n_grid, cv=5, return_train_score=True)

        #fit data
        gridcv.fit(X_trn, y_trn)

        best_param = gridcv.best_params_

        #y_predicted 
        y_pred = gridcv.predict(X_tst)
        print(classification_report(y_tst, y_pred))

        print(50*"-")
        print(f"{str(scalar)}, {str(classifier)}()")
        print(f"[+] Best parameter: {best_param}\n[+] Test Score: {gridcv.score(X_tst, y_tst)}\n[+] Train Score: {gridcv.score(X_trn, y_trn)}\n[+] Total runtime: {datetime.now()-start}")
  
```

```python
from sklearn.pipeline import Pipeline

def create_unscaled_pipeline(_classifier):

    start=datetime.now()



    #excluding ID from model
    features = list(train_dummied.keys())
    features.remove('y_yes')
    features.remove('ID')

    #defining data set
    X_trn = train_dummied.loc[:, features]
    y_trn = train_dummied.loc[:, ['y_yes']]
    X_tst = test_dummied.loc[:, features]
    y_tst = test_response_dummied.loc[:, ['y_yes']]


    if _classifier == "LinearSVC":

        #defining pipeline
        pipe = Pipeline(steps=[('model', LinearSVC())])
        # print(pipe.get_params().keys())


        #get param grid
        c_grid = {"model__C": np.logspace(-3,3,10)}

        #execute GridSearch
        gridcv = GridSearchCV(pipe, c_grid, cv=5, return_train_score=True)
        
        #fit data
        gridcv.fit(X_trn, y_trn)

        #best param
        best_param = gridcv.best_params_

        #y_predicted 
        y_pred = gridcv.predict(X_tst)
        print(classification_report(y_tst, y_pred))


        print(50*"-")
        print(f"No scalar, {str(_classifier)}()")
        print(f"[+] Best parameter: {best_param}\n[+] Test Score: {gridcv.score(X_tst, y_tst)}\n[+] Train Score: {gridcv.score(X_tst, y_tst)}\n[+] Total runtime: {datetime.now()-start}")


    if _classifier == "LogisticRegression":
    
        #defining pipeline
        pipe = Pipeline(steps=[('model', LogisticRegression())])
        #print(pipe.get_params().keys())


        #get param grid
        c_grid = {"model__C": np.logspace(-3,3,10)}

        #execute GridSearch
        gridcv = GridSearchCV(pipe, c_grid, cv=5, return_train_score=True)
        
        #fit data
        gridcv.fit(X_trn, y_trn)

        #best param
        best_param = gridcv.best_params_

        #y_predicted 
        y_pred = gridcv.predict(X_tst)
        print(classification_report(y_tst, y_pred))


        print(50*"-")
        print(f"No scalar, {str(_classifier)}()")
        print(f"[+] Best parameter: {best_param}\n[+] Test Score: {gridcv.score(X_tst, y_tst)}\n[+] Train Score: {gridcv.score(X_tst, y_tst)}\n[+] Total runtime: {datetime.now()-start}")

    if _classifier == "KNN":
        
        #defining pipeline
        pipe = Pipeline(steps=[('model', KNeighborsClassifier())])

        #get param grid
        n_grid = {"model__n_neighbors": range(1,15)}

        #execute GridSearch
        gridcv = GridSearchCV(pipe, n_grid, cv=5, return_train_score=True)
        
        #fit data
        gridcv.fit(X_trn, y_trn)

        #best param
        best_param = gridcv.best_params_

        #y_predicted 
        y_pred = gridcv.predict(X_tst)
        print(classification_report(y_tst, y_pred))


        print(50*"-")
        print(f"No scalar, {str(_classifier)}()")
        print(f"[+] Best parameter: {best_param}\n[+] Test Score: {gridcv.score(X_tst, y_tst)}\n[+] Train Score: {gridcv.score(X_tst, y_tst)}\n[+] Total runtime: {datetime.now()-start}")


```
